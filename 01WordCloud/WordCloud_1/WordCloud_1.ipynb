## WordCloud_1
以英文为例，显示词云制作过程

## 导入词云包

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

## 生成对象

```python
text = open('../constitution.txt').read()
wc = WordCloud().generate(text)
```

## 显示词云

```python
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

## 保存图片

```python
wc.to_file('wordcloud.png')
```
